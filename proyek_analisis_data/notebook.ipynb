## Exploratory Data Analysis (EDA)

### Explore ...

```python
plt.scatter(df_day['temp'], df_day['cnt'], color='blue', alpha=0.6)
plt.title('Temperature vs. Bike Rental Counts')
plt.xlabel('Temperature (temp)')
plt.ylabel('Total Bike Rentals (cnt)')
plt.grid()
plt.show()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
plt.figure(figsize=(10, 6))
plt.bar(df_hour['hr'], df_hour['cnt'], color='orange')
plt.title('Jumlah Penyewaan Sepeda Berdasarkan Jam')
plt.xlabel('Jam dalam Sehari')
plt.ylabel('Jumlah Penyewaan Sepeda (cnt)')
plt.xticks(df_hour['hr'])  # Menampilkan setiap jam di sumbu x
plt.grid(axis='y')  # Menambahkan grid pada sumbu y
plt.show()
```

**Insight:**
- Jumlah penyewaan sepeda mencapai puncaknya sekitar pukul 8 dan 17.
- Terdapat penurunan signifikan dalam jumlah penyewaan antara pukul 0 hingga 6.

### Pertanyaan 2:

```python
# Menghitung jumlah penyewaan berdasarkan musim

# Mengganti nomor musim dengan nama musim
season_mapping = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
df_day['season'] = df_day['season'].map(season_mapping)

# Menghitung total penyewaan sepeda berdasarkan musim
total_rentals_per_season = df_day.groupby('season')['cnt'].sum().reset_index()

# Visualisasi Total Penyewaan Sepeda Berdasarkan Musim
plt.figure(figsize=(10, 6))
bars = plt.bar(total_rentals_per_season['season'], total_rentals_per_season['cnt'], color='orange', alpha=0.7)

# Menambahkan ukuran di atas batang
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval),
             ha='center', va='bottom', fontsize=12)

plt.title('Total Bike Rentals by Season')
plt.xlabel('Season')
plt.ylabel('Total Bike Rentals (cnt)')
plt.grid(axis='y')  # Menambahkan grid pada sumbu y
plt.show()
```

**Insight:**
- Jumlah penyewaan sepeda cenderung lebih tinggi pada musim gugur sekitar 1 juta lebih penyewa.
- lalu turun kembali hampir dengan setengahnya jika memasuki musinm semi dari musim gugur. lalu naik hingga angka naik ke lebih dari 900 ribu penyewa.
